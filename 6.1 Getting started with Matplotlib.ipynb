```python
import matplotlib.pyplot as plt
```

```python
#matplotlib is a collection of command style fn. Each pyplot fn makes some changes to a fig eg: creates a fig, creates a plotting area, plots some lines in a plotting area, decorates the plot with labels etc
```

```python
plt.plot([2,4,6,4]) #list of y values, when x values are not mentioned the indices here 0,1,2,3 become x axis, to pass x and y axis, use 2 lists
plt.ylabel("Numbers")
plt.xlabel("Indices")
plt.title("MyPlot")
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16]) #list of x axis and y axis values
plt.ylabel("Squares")
plt.xlabel("Numbers")
plt.title("MyPlot")
plt.grid() #grid on
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16],'ro') #x axis, y axis, ro = put o-shaped red dots on the points
plt.grid()
plt.show()
```

```python
import numpy as np
t = np.arange(0.,5.,0.2) #start=0, end=5, stepsize=0.2 (0.0, 0.2,0.4,.....4.8)

plt.plot(t,t**2,'b--',label='^2') #x=t, y=t**2,blue dashes
plt.plot(t,t**2.2,'rs',label='^2.2') #x=t, y=t**2.2,red squares
plt.plot(t,t**2.5,'g^',label='^2.5') #x=t, y=t**2.5,green triangles
plt.grid()
plt.legend()
plt.show()
```

```python
#controlling line properties
x = [1,2,3,4]
y=[1,4,9,16]
plt.plot(x,y,linewidth=5.0)
plt.show()
```

```python
#set properties
x1 = [1,2,3,4]
y1 = [1,4,9,16]
x2 = [1,2,3,4]
y2 = [2,4,6,8]
lines = plt.plot(x1,y1,x2,y2) #plots x1 v/s y1 and x2 v/s y2 on the same graph, returns 2 lines as a list

#use kwargs
plt.setp(lines[0],color='r',linewidth=2.0) #set properties on the lines[0] wrt x1, y1

#or matlab style string value pairs
plt.setp(lines[1],'color','g','linewidth',2.0)#set properties on the lines[1] wrt x2, y2

plt.grid()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1) #creates a figure no. 1 where we can create subplots

#subplot command specifies numrows, numcols
#fignum where fignum ranges from 1 to numrows*numcols

plt.subplot(211) #211 means 2 sub plots with 2 rows and 1 column (2=no of rows, 1=no of cols, 1=subplot no, create it in figure 1)
plt.grid() #add grid to the 1st subplot
plt.plot(t1, f(t1), 'b-') #plot in the 1st subplot of fig 1

plt.subplot(212)  #212 means 2 sub plots with 2 rows and 1 column (2=no of rows, 1=no of cols, 2=subplot no, create it in figure 1)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--') #plot in the 2nd subplot of fig 1

plt.show()
```

```python
plt.figure(1) #first figure

plt.subplot(211) #Figure 1 divided into 2 subplots(2 rows, 1 column)
plt.plot([1,2,3]) #plot in figure 1, 1st subplot

plt.subplot(212) #Figure 1 divided into 2 subplots(2 rows, 1 column)
plt.plot([4,5,6]) #plot in figure 1, 2nd subplot

plt.figure(2) #second figure
plt.plot([4,5,6]) #creates a subplot by default of 111 which is 1 row and 1 column

plt.figure(1) #to go back to figure 1
plt.subplot(211) #go back to subplot 211
plt.title("Easy as 1,2,3") #give title to figure 1's subplot 211

plt.show()
```
